```python
p=iris
```

```python
y = p.drop(['sepal_length','petal_length'],axis=1)
y
```

```python
y.plot(legend=True)
```

```python
x = p.drop(['sepal_width','petal_width'],axis=1)
x
```

### secondary_y = True means curve starts from right side

```python
y.plot(figsize=(10,5))
x.plot(figsize=(5,5),secondary_y=True)
```

### Plot sepal_width ,petal_width from left side and sepal_length, petal_length from right side

```python
ax = y.plot(figsize=(10,5))
x.plot(figsize=(10,5),secondary_y=True,ax=ax)
```

### sharex = Flase means every subplot has its own x axis and layout(2,2) represent subplot in form of 2x2 matrix

```python
p.plot(subplots = True,sharex = False,layout=(2,2))
plt.tight_layout()
```
